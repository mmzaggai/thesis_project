```python
import pandas as pd
from pathlib import Path


INPUT_CSV = "topics_per_doc_mts30_fast.csv"  # change if your name is different


df = pd.read_csv(INPUT_CSV, low_memory=False)

print("Columns read:", list(df.columns))  # quick sanity check

out_path = Path(INPUT_CSV).with_name(Path(INPUT_CSV).stem + "_semicolon.csv")
df.to_csv(out_path, index=False, sep=";", encoding="utf-8-sig")

print(f"[ok] Wrote semicolon-separated file → {out_path}")
```
